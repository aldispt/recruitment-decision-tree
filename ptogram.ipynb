```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier, export_text
from sklearn.preprocessing import LabelEncoder

data = {
    'Pendidikan': ['S1', 'D3', 'S2', 'S1', 'D3', 'S2', 'S1', 'S1', 'D3', 'S2', 'S1', 'S1', 'D3', 'S1', 'S2', 'D3', 'S1', 'S2', 'D3', 'S1'],
    'Pengalaman': ['< 2 Thn', '> 5 Thn', '2-5 Thn', '> 5 Thn', '< 2 Thn', '> 5 Thn', '2-5 Thn', '> 5 Thn', '2-5 Thn', '< 2 Thn', '< 2 Thn', '2-5 Thn', '> 5 Thn', '2-5 Thn', '> 5 Thn', '< 2 Thn', '2-5 Thn', '< 2 Thn', '2-5 Thn', '> 5 Thn'],
    'Skill': ['Tinggi', 'Rendah', 'Tinggi', 'Rendah', 'Tinggi', 'Tinggi', 'Rendah', 'Tinggi', 'Tinggi', 'Rendah', 'Rendah', 'Tinggi', 'Tinggi', 'Tinggi', 'Rendah', 'Rendah', 'Tinggi', 'Tinggi', 'Rendah', 'Tinggi'],
    'Wawancara': ['Baik', 'Buruk', 'Baik', 'Buruk', 'Buruk', 'Baik', 'Baik', 'Baik', 'Baik', 'Buruk', 'Baik', 'Buruk', 'Baik', 'Baik', 'Baik', 'Buruk', 'Buruk', 'Baik', 'Baik', 'Buruk'],
    'Hasil': ['Diterima', 'Ditolak', 'Diterima', 'Ditolak', 'Ditolak', 'Diterima', 'Ditolak', 'Diterima', 'Diterima', 'Ditolak', 'Ditolak', 'Ditolak', 'Diterima', 'Diterima', 'Ditolak', 'Ditolak', 'Ditolak', 'Diterima', 'Ditolak', 'Diterima']
}

df = pd.DataFrame(data)

le_pendidikan = LabelEncoder()
le_pengalaman = LabelEncoder()
le_skill = LabelEncoder()
le_wawancara = LabelEncoder()
le_hasil = LabelEncoder()

df['Pendidikan_n'] = le_pendidikan.fit_transform(df['Pendidikan'])
df['Pengalaman_n'] = le_pengalaman.fit_transform(df['Pengalaman'])
df['Skill_n'] = le_skill.fit_transform(df['Skill'])
df['Wawancara_n'] = le_wawancara.fit_transform(df['Wawancara'])
y = le_hasil.fit_transform(df['Hasil'])

X = df[['Pendidikan_n', 'Pengalaman_n', 'Skill_n', 'Wawancara_n']]

clf = DecisionTreeClassifier(criterion='entropy')
clf.fit(X, y)

print(export_text(clf, feature_names=['Pendidikan', 'Pengalaman', 'Skill', 'Wawancara']))

test_input = [[
    le_pendidikan.transform(['S1'])[0],
    le_pengalaman.transform(['< 2 Thn'])[0],
    le_skill.transform(['Tinggi'])[0],
    le_wawancara.transform(['Baik'])[0]
]]

prediksi = clf.predict(test_input)
print(le_hasil.inverse_transform(prediksi))
```
